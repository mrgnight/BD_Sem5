```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
#load dataset
store_data = pd.read_csv('data/dataset123.csv', header = None)
```

```python
#check dataset
store_data
```

```python
#shape check
store_data.shape
```

```python
#convert dataframe into list of lists
records = []
for i in range(0, 22):
    records.append([str(store_data.values[i,j]) for j in range(0, 6)])
```

```python
#first apriori model
association_rules = apriori(records, min_support = 0.50, min_confidence = 0.7, min_lift = 1.2, min_length = 2)
association_rules = list(association_rules)
```

```python
#check number of rules
print(len(association_rules))
```

```python
#check model of rules
print(association_rules)
```
